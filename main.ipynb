```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("coast.png")
imghsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
invr = (imghsv[:,:,0]-np.min(imghsv[:,:,0]))/np.max(imghsv[:,:,0]) #invert

threshold = cv2.threshold(invr,0.29,1,cv2.THRESH_BINARY)[1].astype(np.uint8)
edge = cv2.Canny(threshold,0,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.imshow(edge,cmap="gist_gray",alpha=0.5)
```

```python
img = cv2.imread("foto.jpg")
imghsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
invr = (imghsv[:,:,2]-np.min(imghsv[:,:,2]))/np.max(imghsv[:,:,2]) #invert

threshold = cv2.threshold(invr,0.3,1,cv2.THRESH_BINARY)[1].astype(np.uint8)
blur = cv2.GaussianBlur(threshold,(11,11),3)
edge = cv2.Canny(blur,0,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.imshow(edge,cmap="gist_gray",alpha=0.5)
```
